```python
import pandas as pd
import json
```

## Extract runtimes

```python
columns = ['Original data', 'Number of planes','Number of runways', 'MIP']

rdf = pd.DataFrame(columns=columns)

for i in range(1, 9):
    for j in range(1,3):
        origin_file = f"airland{i}.txt"
        path_1 = f"./Data/{origin_file}_{j}_runway_solution_time.txt"
        path_2 = f"./Data/{origin_file}"
        with open(path_1) as f:
            lines = f.readlines()
            time = float(lines[1].split('=')[1].strip().split(' ')[0])
        with open(path_2) as p:
            line = p.readline()
            planes = int(line.strip().split()[0])
            record = [origin_file, planes, j, time]
            rdf.loc[len(rdf)] = record
          
                
```

## Extract model results

```python
for i in range(1, 9):
    for j in range(1,3):
        origin_file = f"airland{i}.txt"
        path = f"./Data/{origin_file}_{j}runaway_solution.json"
        with open(path) as f:
            data = json.load(f)
        result = data['CPLEXSolution']['variables']
        if i == 1 and j == 1:
            df = pd.DataFrame.from_records(result)
            df['Number of runways'] = j
            df['Original data'] = origin_file
        else:
            df_n = pd.DataFrame.from_records(result)
            df_n['Number of runways'] = j
            df_n['Original data'] = origin_file
            
            result = pd.concat([df, df_n])
            
result  
```
